# Normal Probability Plots

## Key Concepts:

### Percentile:
A **percentile** is the value below which a certain percentage of the data in a dataset falls.  
**Example**: The 25th percentile (also called the first quartile) is the value below which 25% of the data lies.

### Quantiles:
A **quantile** is a generalization of percentiles. Instead of dividing data into 100 parts like percentiles, quantiles divide data into any number of equal parts.
**Example**: The 0.25 quantile is equivalent to the 25th percentile.

### Z-scores:
A **Z-score** represents the standardized value of a dataset. It tells us how far (in terms of standard deviations) a value is from the mean.

---

## Normal Probability Plot

The **normal probability plot** is a specific type of Q-Q plot used to assess whether a dataset follows a normal distribution. In this plot, the ordered sample data is plotted against the theoretical quantiles from a normal distribution.


The normal probability plot is used to answer the following questions:
* Is Data normally distributed?If yes,points lie close to straight line.
* If not, what is the nature of distribution?Skewness


```python
#Import dependencies
from scipy.stats import zscore,norm
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import statsmodels.graphics.gofplots as sm
```

```python

#its just an example npp of original values v/s theortical values(z scores)
def npp(data):
    data = sorted(data)
    p = [(data.index(i)-0.5)/len(data) for i in data]
    z = zscore(p)

    plt.scatter(data, z)
    plt.ylabel('Z values')
    plt.xlabel(' values')
    plt.show()
    plt.plot(data,  z,'ro',data,  data)
    plt.show()
```

```python
sample_size=10000

standard_norm_data = np.random.randn(sample_size)
print(standard_norm_data)
npp(standard_norm_data)

# Plot the bell curve
sns.histplot(standard_norm_data,kde=True)
```
